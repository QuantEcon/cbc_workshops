A monopolist faces inverse demand
curve

$$    P_t = a_0 - a_1 Y_t + Z_t, $$

where 

* $P_t$ is price, 
* $Y_t$ is output and 
* $Z_t$ is a demand shock.

We assume that $Z_t$ is a discretized AR(1) process.

Current profits are 

$$ P_t Y_t - c Y_t - \gamma (Y_{t+1} - Y_t)^2 $$

Combining with the demand curve and writing $y, y'$ for $Y_t, Y_{t+1}$, this becomes

$$    r(y, z, y′) := (a_0 - a_1  y + z - c) y - γ  (y′ - y)^2 $$

The firm maximizes present value of expected discounted profits.  The Bellman equation is 

$$   v(y, z) = \max_{y'} \left\{ r(y, z, y′) + β \sum_{z′} v(y′, z′) Q(z, z′) \right\}. $$

We discretize $y$ to a finite grid `y_grid`.

In essence, the firm tries to choose output close to the monopolist profit maximizer, given $Z_t$, but is constrained by adjustment costs.

We will use the following imports.

```python
from collections import namedtuple
import numpy as np
import quantecon as qe
from numba import njit, prange, int32
import matplotlib.pyplot as plt
```

```python
@njit
def argmax(list_object):
    "Return the index of the largest element of `list_object`."
    max_val = -np.inf
    argmax_index = None
    for i, x in enumerate(list_object):
        if x > max_val:
            max_val = x
            argmax_index = i
    return argmax_index
```

We also need the successive approximation function.

```python
def successive_approx(T,                     # Operator (callable)
                      x_0,                   # Initial condition
                      tolerance=1e-6,        # Error tolerance
                      max_iter=10_000,       # Max iteration bound
                      print_step=25,         # Print at multiples
                      verbose=False):        
    x = x_0
    error = tolerance + 1
    k = 1
    while error > tolerance and k <= max_iter:
        x_new = T(x)
        error = np.max(np.abs(x_new - x))
        if verbose and k % print_step == 0:
            print(f"Completed iteration {k} with error {error}.")
        x = x_new
        k += 1
    if error > tolerance:
        print(f"Warning: Iteration hit upper bound {max_iter}.")
    elif verbose:
        print(f"Terminated successfully in {k} iterations.")
    return x
```

## Primitives and operators

We used a `namedtuple` to store parameters and grids for the state space:

```python

Model = namedtuple("Model", 
                   ("β", "a_0", "a_1", "γ", "c",
                    "y_size", "z_size", "y_grid", "z_grid", "Q"))

def create_investment_model(
        r=0.01,                              # Interest rate
        a_0=10.0, a_1=1.0,                   # Demand parameters
        γ=25.0, c=1.0,                       # Adjustment and unit cost 
        y_min=0.0, y_max=20.0, y_size=100,   # Grid for output
        ρ=0.9, ν=1.0,                        # AR(1) parameters
        z_size=150):                         # Grid size for shock
    """
    A function that takes in parameters and returns an instance of Model that
    contains data for the investment problem.
    """
    β = 1/(1+r) 
    y_grid = np.linspace(y_min, y_max, y_size)  
    mc = qe.tauchen(ρ, ν, n=z_size)
    z_grid, Q = mc.state_values, mc.P

    model = Model(β=β, a_0=a_0, a_1=a_1, γ=γ, c=c,
                  y_size=y_size, z_size=z_size,
                  y_grid=y_grid, z_grid=z_grid, Q=Q)
    return model
```

The function below is the right-hand side of the Bellman equation (unmaximized)

```python
@njit
def B(i, j, ip, v, model):
    """
    The right-hand side of the Bellman equation with candidate value v and
    arguments y[i], z[j], yp[ip].

    """
    β, a_0, a_1, γ, c, y_size, z_size, y_grid, z_grid, Q = model
    y, z, yp = y_grid[i], z_grid[j], y_grid[ip]
    r = (a_0 - a_1 * y + z - c) * y - γ * (yp - y)**2
    return r + β * np.dot(v[ip, :], Q[j, :]) 
```

We define the policy and Bellman operators, as well as a function to compute greedy policies.

```python
@njit(parallel=True)
def T_σ(v, σ, model):
    "The policy operator."
    v_new = np.empty_like(v)
    for i in prange(model.y_size):
        for j in range(model.z_size):
            v_new[i, j] = B(i, j, σ[i, j], v, model)
    return v_new


@njit(parallel=True)
def T(v, model):
    "The Bellman operator."
    β, a_0, a_1, γ, c, y_size, z_size, y_grid, z_grid, Q = model
    v_new = np.empty_like(v)
    for i in prange(y_size):
        for j in range(z_size):
            v_new[i, j] = max([B(i, j, ip, v, model) for ip in range(y_size)])
    return v_new


@njit(parallel=True)
def get_greedy(v, model):
    "Compute a v-greedy policy."
    β, a_0, a_1, γ, c, y_size, z_size, y_grid, z_grid, Q = model
    σ = np.empty_like(v, dtype=int32)
    for i in prange(y_size):
        for j in range(z_size):
            σ[i, j] = argmax([B(i, j, ip, v, model) for ip in range(y_size)])
    return σ
```

The next function computes the lifetime value of following policy $\sigma$.

The basic problem is to solve the linear system

$$ v(y, z) = r(y, z, \sigma(y, z)) + β \sum_{z'} v(\sigma(y, z), z') Q(z, z) $$

for $v$.  

It turns out to be helpful to rewrite this as 

$$ v(y, z) = r(y, z, \sigma(y, z)) + β \sum_{y', z'} v(y', z') P_\sigma(y, z, y', z') $$

where $P_\sigma(y, z, y', z') = 1\{y' = \sigma(y, z)\} Q(z, z')$.

We want to write this as $v = r_\sigma + P_\sigma v$ and then solve for $v$

Note, however, that $v$ is a multi-index array, rather than a vector.  

The code below 

1. reshapes $v$ and $r_\sigma$ to 1D arrays and $P_\sigma$ to a matrix
2. solves the linear system
3. converts back to multi-index arrays.

```python
@njit(parallel=True)
def get_value(σ, model):
    "Get the value v_σ of policy σ."
    # Unpack and set up
    β, a_0, a_1, γ, c, y_size, z_size, y_grid, z_grid, Q = model
    ny, nz = len(y_grid), len(z_grid)
    n = ny * nz
    # Allocate and create multi-index versions of P_σ and r_σ
    P_σ = np.zeros((ny, nz, ny, nz))
    r_σ = np.zeros((ny, nz))
    for i in range(ny):
        for j in range(nz): 
            y, z, yp = y_grid[i], z_grid[j], y_grid[σ[i, j]]
            r_σ[i, j] = (a_0 - a_1 * y + z - c) * y - γ * (yp - y)**2
            for ip in range(ny):
                for jp in range(nz): 
                    if ip == σ[i, j]:
                        P_σ[i, j, ip, jp] = Q[j, jp]
    # Reshape to standard matrix algebra form
    r_σ = r_σ.reshape((n,))
    P_σ = P_σ.reshape((n, n))
    # Solve for the value of σ 
    I = np.identity(n)
    v_σ = np.linalg.solve(I - β * P_σ, r_σ)
    # Return as multi-index array
    return np.reshape(v_σ, (ny, nz))
```


Next we define three solvers, which implement VFI, HPI and OPI.

```python

def value_iteration(model, tol=1e-5):
    "Implements VFI."
    vz = np.zeros((len(model.y_grid), len(model.z_grid)))
    v_star = successive_approx(lambda v: T(v, model), vz, tolerance=tol)
    return get_greedy(v_star, model)

def policy_iteration(model):
    "Howard policy iteration routine."
    ny, nz = len(model.y_grid), len(model.z_grid)
    σ = np.ones((ny, nz), dtype=int)
    i, error = 0, 1.0
    while error > 0:
        v_σ = get_value(σ, model)
        σ_new = get_greedy(v_σ, model)
        error = np.max(np.abs(σ_new - σ))
        σ = σ_new
        i = i + 1
        print(f"Concluded loop {i} with error {error}.")
    return σ

def optimistic_policy_iteration(model, tol=1e-5, m=100):
    "Implements the OPI routine."
    v = np.zeros((len(model.y_grid), len(model.z_grid)))
    error = tol + 1
    while error > tol:
        last_v = v
        σ = get_greedy(v, model)
        for _ in range(m):
            v = T_σ(v, σ, model)
        error = np.max(np.abs(v - last_v))
    return get_greedy(v, model)
```

Here's a more systematic test, with plots.

## Exercise

If $\gamma=0$, then there is no intertemporal trade off and the profit maximizing level of output is 

$$ \bar Y_t := (a_0 - c + Z_t) / (2  a_1) $$

The optimal policy is to choose $\bar Y_t$ in every period.

If $\gamma > 0$, then the firm should implement some smoothing, to take into account adjustment costs.

Try to show, in a time series simulation, how this smoothing is implemented by the firm, with more smoothing for larger values of $\gamma$.

```python
for _ in range(14):
    print('solution below')
```

```python
ts_length = 200
fig, axes = plt.subplots(4, 1, figsize=(9, 11.2))

for (ax, γ) in zip(axes, (1, 10, 20, 30)):
    model = create_investment_model(γ=γ)
    β, a_0, a_1, γ, c, y_size, z_size, y_grid, z_grid, Q = model
    σ_star = optimistic_policy_iteration(model)
    mc = qe.MarkovChain(Q, z_grid)

    z_sim_idx = mc.simulate_indices(ts_length)
    z_sim = z_grid[z_sim_idx]
    y_sim_idx = np.empty(ts_length, dtype=int)
    y_1 = (a_0 - c + z_sim[0]) / (2 * a_1)
    y_sim_idx[0] = np.searchsorted(y_grid, y_1)
    for t in range(ts_length-1):
        y_sim_idx[t+1] = σ_star[y_sim_idx[t], z_sim_idx[t]]
    y_sim = y_grid[y_sim_idx]
    y_bar_sim = (a_0 - c + z_sim) / (2 * a_1)

    ax.plot(range(ts_length), y_sim, label="$Y_t$")
    ax.plot(range(ts_length), y_bar_sim, label="$\\bar Y_t$")
    ax.legend(fontsize=fontsize, frameon=False, loc="upper right")
    ax.set_ylabel("output", fontsize=fontsize)
    ax.set_ylim(1, 9)
    ax.set_title(f"$\gamma = {γ}$", fontsize=fontsize)

fig.tight_layout()
plt.show()
```
